```python
import numpy as np
import pandas as pd

mydata = pd.read_csv("./USA_coviddata_clean_25.csv")
mydata.index = range(mydata.shape[0])
#take useless columns aside. It will be used to concatenate in the end
someinfo = mydata.iloc[:,0:6]
mydata = mydata.drop(labels=['sub_region_1','date','hospitalized_cumulative','hospitalized_new','Unnamed: 0'],axis=1)
#get the median of each symtpom region pair
median_across_symptoms = mydata.groupby('open_covid_region_code').agg('median')
print(median_across_symptoms)
#print(region_median)
```

```python
#median of each state
median_median = median_across_symptoms.agg('median',axis=1)

print(median_median)
```

```python

for index, row in mydata.iterrows():
    
    row[1:] = row[1:]/median_median[row['open_covid_region_code']]

    mydata.iloc[index,1:] = row[1:]
    #print(mydata.iloc[index,:])

```

```python
result = pd.concat([someinfo,mydata.iloc[:,1:]],axis=1)
result.to_csv('./demedian_USA_coviddata_clean_25.csv', index=False)
```
