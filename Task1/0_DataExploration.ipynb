```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd

df = pd.read_csv('../Data/aggregated_cc_by_sa.csv', low_memory=False)
df_USAstates =  df.loc[df['open_covid_region_code'].str.contains('US-')]
# Extract Region Code For Plotting
df_USAstates.loc[:,'open_covid_region_code'] = df_USAstates['open_covid_region_code'].str.replace('US-','')
# df_USAstates['hospitalized_new']
```

```python
df_USAstates = df_USAstates.sort_values(by=['date']) 
df_USAstates_clean = df_USAstates[['open_covid_region_code','region_name','date',
                                  'hospitalized_cumulative','hospitalized_new']]
states = df_USAstates_clean['region_name'].unique()

# Remove States with > XX consecutive days of 0.0 values
count = 1
df_final = df_USAstates_clean.copy()
for state in states:
    df_state = df_USAstates_clean.loc[df_USAstates_clean['region_name']==state]
    print(count,state, df_state.shape[0], df_state['hospitalized_new'].value_counts()[0.0],
         df_state['hospitalized_cumulative'].value_counts()[0.0])
    count+=1
    if df_state['hospitalized_new'].value_counts()[0.0] > df_state.shape[0]*0.8:
        print(state + " has been removed")
        df_final = df_final[(df_final.region_name != state)]
    
    
```

```python
fig = px.choropleth(df_final, 
                    locations="open_covid_region_code",
                    color="hospitalized_new", # lifeExp is a column of gapminder
                    hover_name="region_name", # column to add to hover information
                    color_continuous_scale = 'reds',
                    animation_frame="date",
                    range_color=(0, 10),
                    scope = "usa",
                    locationmode="USA-states"
                   )
fig.show()

## Using go plots
# dates = df.date.unique()
# for date in dates: 
#     df_plot = df.loc[df['date'] == date]
#     fig = go.Figure(data=go.Choropleth(
#         locations=df_plot['open_covid_region_code'], # Spatial coordinates
#         z = df_plot['cases_cumulative'].astype(float), # Data to be color-coded
#         colorscale = 'Reds',
#         colorbar_title = "Cumulative Cases",
#     ))

#     fig.update_layout(
#         title_text = '',
#     )
#     fig.show()
```
